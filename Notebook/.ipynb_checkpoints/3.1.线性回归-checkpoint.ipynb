# 矢量化加速

在训练我们的模型时，我们经常希望能够同时处理整个小批量的样本。 为了实现这一点，需要我们对计算进行矢量化， 从而利用线性代数库，而不是在Python中编写开销高昂的for循环。

```python
import math
import time
import numpy as np
import torch
from d2l import torch as d2l

n = 1000000
a=torch.ones(n).reshape(1000,1000)
b=torch.ones(n).reshape(1000,1000)
```

```python
class Timer:
    # 记录多次运行时间
    def __init__(self):
        self.times = []
        self.start()

    def start(self):
        self.tik = time.time()

    def stop(self):
        self.times.append(time.time() - self.tik)
        return self.times[-1]

    def avg(self):
        return sum(self.times)/len(self.times)

    def sum(self):
        return sum(self.times)

    def cumsum(self):
        return np.array(self.times).cumsum().tolist()
```

对负载进行基准测试

```python
c = torch.zeros(n).reshape(1000,1000)
timer = Timer()
for i in range(0,1000):
    for j in range(0,1000):
        c[i][j] = a[i][j] + b[i][j]

print(f'{timer.stop():.5f} sec')
```

```python
timer.start()
d = a + b
print(f"{timer.stop():.5f}")
```

可以看出，矢量化代码会带来数量级的加速

# 正态分布与平方损失

正态分布也称为高斯分布，由德国数学家高斯应用于天文研究。

它描述的是一组数据集中，大部分数据值集中在中间，远离中间的数值则越来越少。

参数：
1. 均值：正态分布的中心点，通常表示为 μ。均值决定了分布的“中心位置”，就是钟形曲线的顶端。
2. 标准差：衡量数据分布的宽窄，通常表示为 σ。标准差越大，钟形曲线越宽，数据分布得越分散；标准差越小，曲线越尖，数据分布得越集中。

其概率密度函数如下：



```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

x = np.arange(-7, 7, 0.01)
# 均值和标准差对
params = [(0, 1), (0, 2), (3, 1)]
d2l.plot(x, [normal(x, mu, sigma) for mu, sigma in params], xlabel='x',
         ylabel='p(x)', figsize=(4.5, 2.5),
         legend=[f'mean {mu}, std {sigma}' for mu, sigma in params])
```

改变均值会产生沿轴的偏移，增加方差将会分散分布、降低其峰值。
