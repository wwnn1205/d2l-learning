# 分类问题

在计算机中如何给图片打标签?有两种方法:

1. 数字标签
这种方法最直观，就是用数字来代表不同的类别。比如：
+ 1代表猫
+ 2代表狗
+ 3代表鸡

这种方法虽然简单，但有个问题：如果把“1”和“2”放在一起，电脑可能会以为它们之间有某种大小关系，比如“狗”比“猫”大，或者“猫”比“狗”小。但实际上，“猫”“狗”“鸡”这三个类别之间并没有这种大小或顺序关系。

2. 独热编码(One-hot Encoding)

这是一种更常用、更“安全”的方法。它不用一个数字来代表一个类别，而是用一个向量（也就是一串数字）来代表。这个向量里只有一个位置是“1”，其他位置都是“0”。

+ 如果图片是猫，标签就是 (1, 0, 0)。
+ 如果图片是狗，标签就是 (0, 1, 0)。
+ 如果图片是鸡，标签就是 (0, 0, 1)。

通过这种方式，电脑就不会误以为不同类别之间有顺序关系了。1 所在的位置，就是唯一的正确答案。

神经网络通过将输入的特征（x）乘以各自的重要性（w）并相加，来计算出每一个类别的预测值（o），从而完成分类任务。





我们可以用神经网络图 <a href="#fig3.4.1">图3.4.1</a> 来描述这个计算过程。与线性回归一样，softmax回归也是一个单层神经网络。由于计算每个输出$o_1$、$o_2$和$o_3$取决于所有输入$x_1$、$x_2$、$x_3$和$x_4$，所以softmax回归的输出层也是全连接层。

为了更简洁地表达模型，我们仍然使用线性代数符号。通过向量形式表达为$\mathbf{o} = \mathbf{W}\mathbf{x} + \mathbf{b}$，这是一种更适合数学和编写代码的形式。由此，我们已经将所有权重放到一个$3 \times 4$矩阵中。对于给定数据样本的特征$\mathbf{x}$，我们的输出是由权重与输入特征进行矩阵-向量乘法再加上偏置$\mathbf{b}$得到的。

# softmax运算

它的作用是把神经网络的输出结果，变成我们能看懂的**概率**。

神经网络计算出的值$o_1$、$o_2$和$o_3$,他们加起来不一定是1,而且每个值还可能有负数

但是我们想要的是概率,概率要求每一个值的和加起来是1,并且不能有负数

所以用softmax进行转换,将结果转换为符合概率规则的数字

步骤:
1. 指数化:把每个输出o变成e的o次幂,即o=$e^o$,这样做的好处是,所有的结果都会变成正数

2. 归一化:把所有指数化的结果加起来,然后用每一个结果除以这个总和
